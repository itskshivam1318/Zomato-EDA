```python
df_country = pd.read_excel('../input/zomato-restaurants-data/Country-Code.xlsx')
# ,encoding='latin-1')
# pd.read_csv('../input/zomato-restaurants-data/zomato.csv',encoding='latin-1')
df_country.head()
```

```python
print(df_country.columns)
print(df.columns)
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_count = final_df.Country.value_counts().values
country_count
```

```python
# plot piechart

plt.pie(country_count,labels=country_names,autopct='%1.2f%%')
plt.show()
```

```python
# top three country
plt.pie(country_count[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y= "Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y= "Rating Count", data=ratings,hue='Rating color',palette=['black','red','orange','yellow','green','green'])
```

```python
#count plot
# Frequency of the values
sns.countplot(x="Rating color",data=ratings,palette=['black','red','orange','yellow','green','green'])
```

```python
sns.countplot(x="Rating text",data=ratings,palette=['black','red','orange','yellow','green','green'])
```

```python
rating_category = final_df.groupby(['Rating text','Rating color']).size().reset_index().rename(columns={0:'Rating Count'})
rating_category
```

```python
sns.barplot(x='Rating text',y='Rating Count',data=rating_category,hue='Rating color',palette=['orange','green','yellow','black','red','green'])
```
